```python
import os
import numpy as np
from skimage import io
from shutil import copyfile
train_list = os.listdir('stage1_train')
train_list = [x for x in train_list if not x.startswith('.')]
```

```python
train_list = [train_list[0]]
```

```python
#generate cumulative masks
for ii,f in enumerate(train_list):
    print f
    small_masks = os.listdir('stage1_train/'+f+'/masks')
    og_image = os.listdir('stage1_train/'+f+'/images')
    big_mask = 'placeholder'
    for s in small_masks:
        a=io.imread('stage1_train/'+f+'/masks/'+s)
        if big_mask == 'placeholder':
            big_mask = a
        else:
            big_mask = big_mask + a            
    io.imsave('train/y/'+'mask_'+str(ii)+'.png',big_mask)
    copyfile('stage1_train/'+f+'/images/'+og_image[0],'train/x/'+'img_'+str(ii)+'.png')
```

```python
#extract 100 random patches from each image
from sklearn.feature_extraction import image
np.random.seed(0)

og_x_list = os.listdir('train/x/')
og_x_list = [ x for x in og_x_list if not x.endswith('.DS_Store') ]

for x in [og_x_list[0]]:
    img_x = io.imread('train/x/'+x)
    img_y = io.imread('train/y/'+'mask_'+x[4:])
    
    img = np.dstack((img_x,img_y))
    
    print "DSTACK SHAPE: " + str(np.shape(img))
    
    patches = image.extract_patches_2d(img, (100+np.random.randint(low=0,high=100), 100+np.random.randint(low=0,high=100)),100,0) #50% of possible patches, random seed = 0
    dims = np.shape(patches)
    print "PATCHES SHAPE: " + str(dims)
    for i in range(dims[0]):
        mask = patches[i,:,:,-1]
        print "MASK SHAPE: " + str(mask.shape)
        io.imsave('train/y/'+'mask_'+x[4:-4]+'_'+str(i)+'.png',mask)
        
        patch = patches[i,:,:,0:dims[-1]-2]
        print "PATCH SHAPE: " + str(patch.shape)
        io.imsave('train/x/'+'img_'+x[4:-4]+'_'+str(i)+'.png',patch)
```

```python
#rotate and flip

xlist = os.listdir('./train/x/')
xlist = [ x for x in xlist if not x.endswith('.DS_Store') ]
ylist = os.listdir('./train/y/')
ylist = [ x for x in ylist if not x.endswith('.DS_Store') ]

for f in xlist:
    im = io.imread('./train/x/'+f)
    im1 = np.rot90(im)
    im2 = np.rot90(im1)
    im3 = np.rot90(im2)
    f=f[:-4]
    io.imsave('./train/x/'+f+'_r1.png',im1)
    io.imsave('./train/x/'+f+'_r2.png',im2)
    io.imsave('./train/x/'+f+'_r3.png',im3)
    
for f in ylist:
    im = io.imread('./train/y/'+f)
    im1 = np.rot90(im)
    im2 = np.rot90(im1)
    im3 = np.rot90(im2)
    f=f[:-4]
    io.imsave('./train/y/'+f+'_r1.png',im1)
    io.imsave('./train/y/'+f+'_r2.png',im2)
    io.imsave('./train/y/'+f+'_r3.png',im3)
    
xlist = os.listdir('./train/x/')
xlist = [ x for x in xlist if not x.endswith('.DS_Store') ]
ylist = os.listdir('./train/y/')
ylist = [ x for x in ylist if not x.endswith('.DS_Store') ]

for f in xlist:
    im = io.imread('./train/x/'+f)
    im1 = np.flipud(im)
    f=f[:-4]
    io.imsave('./train/x/'+f+'_flip.png',im1)
    
for f in ylist:
    im = io.imread('./train/y/'+f)
    im1 = np.flipud(im)
    f=f[:-4]
    io.imsave('./train/y/'+f+'_flip.png',im1)
```
